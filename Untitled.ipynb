```python
# %load Map.py
# Plot GPS coordinates of photos taken while in Europe

# import libraries
import folium
from folium.plugins import MarkerCluster
import pandas
import glob
import exifread
import base64

# iterate through photos and extract the names and GPS coordinates
list_of_files = glob.glob("Pics/Geotagged/*.jpg")
names = []
lat = []
lat_name = "GPS GPSLatitude"
lon = []
lon_name = "GPS GPSLongitude"

for file in list_of_files:
	f = base64.b64encode(open(file,"rb").read())
	names.append(f.name.split("/")[-1])
	tags = exifread.process_file(f)
	#print(tags.keys())
	if lat_name in tags:
		# convert to decimal
		lat_dec = float(tags[lat_name].values[0].num)/float(tags[lat_name].values[0].den) + float(tags[lat_name].values[1].num)/float(tags[lat_name].values[1].den)/60 + float(tags[lat_name].values[2].num)/float(tags[lat_name].values[2].den)/3600
		lon_dec = float(tags[lon_name].values[0].num)/float(tags[lon_name].values[0].den) + float(tags[lon_name].values[1].num)/float(tags[lon_name].values[1].den)/60 + float(tags[lon_name].values[2].num)/float(tags[lon_name].values[2].den)/3600
		lat.append(lat_dec)
		lon.append(lon_dec)

# create map
Map = folium.Map(location=[47.656999,5.779075],zoom_start=5,tiles="cartodbpositron")

marker_cluster = MarkerCluster().add_to(Map)

# create features for data points
#fgp = folium.FeatureGroup(name="LocPics")
		
# add features to fgp
for lt,ln,nm in zip(lat,lon,names):
	# create html tag
	html_text = r"""<img src=Pics/Geotagged/%s;base64 alt="Mountain View" width=400 max-height = auto """ % nm
	iframe = folium.Html(data=html_text,script=True)
	popup = folium.Popup(html=iframe,parse_html=True, max_width=2650)
	# add child to feature group
	#fgp.add_child(folium.CircleMarker(location=[lt,ln],radius=5,fill=True,popup=popup).add_to(marker_cluster))
	folium.CircleMarker(location=[lt,ln],radius=5,fill=True,popup=popup).add_to(marker_cluster)

# add features to group
#Map.add_child(fgp)


# save map
Map.save("Map_Europe.html")
```
